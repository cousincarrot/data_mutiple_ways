```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
matches = pd.read_csv('world_cup_matches.csv')
matches = matches.rename(columns=lambda x: x.strip())
#loads data, cleans whitespace
```

```python
matches.rename(columns={'Year':'year',
                       'Attendance':'attendance',
                       'Home Team Goals':'home_team_goals',
                       'Away Team Goals':'away_team_goals',
                       'Half-time Home Goals':'half_home_goals',
                       'Half-time Away Goals':'half_away_goals',
                       'Home Team Initials':'home_initials',
                       'Away Team Initials':'away_initials'}, inplace=True)
#renames column
```

```python
matches.head(8)
```

```python
atten = matches['attendance'].dropna()
#drop rows that are NaN
```

```python
atten = atten.apply(lambda x: int(float(x)))
```

#### 1. Choose one variable and plot that variable four different ways: Attendance

```python
#1-1
fig, ax = plt.subplots(figsize=(20,15))
plt.title('Distribution With Markers of Each Data Entry for Attendance', loc='center', fontdict={'fontsize': 15, 'weight':'bold'})
sns.distplot(atten, ax=ax, rug=True);
#PROS: Delinates variable with 3 visuals in one graph
#CONS: I have no idea why the Y value is so small
```

```python
#1-2
fig, ax = plt.subplots(figsize=(20,15))
plt.title('Box Plot of All Matches Attendance', loc='center', fontdict={'fontsize': 15, 'weight':'bold'})
sns.boxplot(y=atten, ax=ax);
#PROS: Clearly see the IQR
#CONS: It's just a piece of the puzzle, it needs supporting graphs.
```

```python
#1-3
fig, ax = plt.subplots(figsize=(20,15))
plt.title('Violin Plot of All Matches Attendance', loc='center', fontdict={'fontsize': 15, 'weight':'bold'})
sns.violinplot(y=atten, ax=ax);
#PROS: More detail in the mode (outline) compared to a box plot. Shows full distribution of the data,
#good for multiple peaks.
#CONS: I think this is meant to have another inner shape to represent the 95th percentile,
#but perhaps it is so small it reads as a line?
```

```python
#1-4
fig, ax = plt.subplots(figsize=(20,15))
plt.title('Count of Attendance of 100 Data Entries', loc='center', fontdict={'fontsize': 15, 'weight':'bold'})
plt.xticks(rotation=90)
sns.countplot(atten[:100], ax=ax);
#PROS: The colors and the legibility of the visuals in genreal (excluding x labels in this instance)
#CONS: Probably not the ideal graph for this databecause there are so many entries, the attendance numbers are not legible.
# Also, the count of number of times the attendance hit the same exact amount of people is not very valuabe information.
```

```python
#1-5 *bonus*
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(15,10))
plt.title('Avg and Standard Devation of Attendance at All Matches', loc='center', fontdict={'fontsize': 15, 'weight':'bold'});
sns.barplot(y=atten, errcolor='orange', ax=ax);
#PROS: Very clear representation of mean (ehight of bar) and standard deviation (line in top)
#CONS: Not that much information given, really. In this case, I would want smaller increments for the Y axis
```

#### 2. Choose two continuous variables, and plot them three different ways.

```python
#2-1a
sns.jointplot(matches['home_team_goals'], matches['away_team_goals'], kind='reg', size=10)
plt.title('Comparing Home and Away Team Goals', loc='right', fontdict={'fontsize': 15, 'weight':'bold'});
```

```python
#2-1b
g = sns.jointplot('home_team_goals', 'away_team_goals', data=matches, kind='kde', space=0, color='g', size=10)
plt.title('Comparing Home and Away Team Goals', loc='right', fontdict={'fontsize': 15, 'weight':'bold'});
```

```python
#2-1c
g = (sns.jointplot('home_team_goals',
                   'away_team_goals',
                    data=goals, color="k")
                    .plot_joint(sns.kdeplot, zorder=0, n_levels=4))
```

```python
#2-2
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(20,15))
sns.boxplot(data=goals)
plt.title('Mean, IQR, and Outliers of Goals for Home and Away Teams at Full Time and Half Time', fontdict={'fontsize': 15, 'weight':'bold'})
```

```python
#2-3
sns.lmplot(x='home_team_goals', y='away_team_goals', data=goals, x_estimator=np.mean, size=10)
plt.title('Comparing Home Team Goals to Away Team Goals', fontdict={'fontsize': 15, 'weight':'bold'});
```

```python
#3-2
# Palette
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(home_50_sort, row='home_initials', hue='home_initials', aspect=15, size=.5, palette=pal)

# Draw the densities in a few steps
g.map(sns.kdeplot, 'home_team_goals', clip_on=False, shade=True, alpha=1, lw=1.5, bw=.2)
g.map(sns.kdeplot, 'home_team_goals', clip_on=False, color='w', lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2, clip_on=False)

# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight='bold', color=color, 
            ha='left', va='center', transform=ax.transAxes)

g.map(label, 'home_team_goals')

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=0)

# Remove axes details that don't play will with overlap
g.set_xlabels('Home Team Goals')
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True);
```

```python
top_9 = home_team_pos[:15].groupby('home_initials').mean().reset_index()
#Find the top 9 scoring home teams by their goal average
```

```python
top_9
```

```python
#Setting the top 9 teams to a new df
top_sums = home_team_pos[home_team_pos['home_initials'].isin(['ARG','BRA','CHI','FRA','PAR','ROU','URU','USA','YUG'])]
top_sums = top_sums.groupby('home_initials').sum().reset_index()
```

```python
top_sums = top_sums.sort_values('home_team_goals', axis=0, ascending=False)
```

```python
top_sums
```

```python
#converting a sweet seabrn color palette to hex so the plt.pie can read them. Pass in name of set, # of results wanted
pal = sns.color_palette("Set3", 9)
pal.as_hex()
```

```python
#3-6
plt.figure(figsize=(6,6))
plt.title('Percentage of Goals Scored Within the Top 9 Scoring Home Teams', fontdict={'fontsize': 15, 'weight':'bold'});
plt.pie(top_sums.home_team_goals,
        labels=top_9.home_initials,
        autopct='%.0f%%',
        pctdistance=0.85,
        explode=(0.05, 0, 0, 0, 0, 0, 0, 0, 0),
        center=(0,10),
        colors=pal.as_hex(),
        labeldistance=1.1,
        radius=1,
        shadow=True,
        startangle=225,
        counterclock=False,
        wedgeprops = {'linewidth': 1.5, 'ec': '#999999'},
        textprops = {'fontname':'Sans', 'fontsize' : 13, 'weight':'normal'}
);
```

```python
goals_melt = pd.melt(frame=home_team_pos, id_vars='home_initials', value_vars='home_team_goals')
#This function stacks the data in a different way...
```

```python
goals_melt
```

```python
goals_melt.value.max()
#Finding the top value of goals_melt
```

```python
goals_melt = goals_melt[goals_melt['value'] > 3]
#restricting the rows to those with over '3' in the goals aka 'value' column of goals_melt
```

```python
# This visualization is just a test of this format.
# The visualization is not a good match for the data, but I wanted to try it out.
f, ax = plt.subplots(figsize=(20, 15))
sns.despine(bottom=True, left=True)

# Show each observation with a scatterplot
sns.stripplot(x='value', y='variable', hue='home_initials',
              data=goals_melt, dodge=True, jitter=True,
              alpha=.25, zorder=1)

# Show the conditional means
sns.pointplot(x='value', y='variable', hue='home_initials',
              data=goals_melt, dodge=.532, join=False, palette="dark",
              markers="d", scale=.75, ci=None)

# Improve the legend 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[3:], labels[3:], title='Home Team Goals',
          handletextpad=0, columnspacing=1,
          loc="lower right", ncol=3, frameon=True)
```
